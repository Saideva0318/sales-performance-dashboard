# Sales Performance EDA
Exploratory Data Analysis for Sales Performance Dashboard

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

pd.set_option('display.float_format', '{:.2f}'.format)
df = pd.read_csv('../data/processed/sales_processed.csv', parse_dates=['date'])
print(f'Shape: {df.shape}')
df.head()
```

```python
# Data Quality Check
print('Missing values:')
print(df.isnull().sum())
print('\nData types:')
print(df.dtypes)
```

```python
# Revenue Distribution
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.histplot(df['revenue'], bins=50, kde=True, color='steelblue')
plt.title('Revenue Distribution')
plt.subplot(1,2,2)
sns.boxplot(data=df, x='category', y='revenue', palette='Set2')
plt.xticks(rotation=45)
plt.title('Revenue by Category')
plt.tight_layout()
plt.show()
```

```python
# Monthly Revenue Trend
monthly = df.groupby('year_month')['revenue'].sum().reset_index()
fig = px.line(monthly, x='year_month', y='revenue', markers=True, title='Monthly Revenue Trend')
fig.show()
```

```python
# Correlation Heatmap
numeric_cols = ['unit_price', 'quantity', 'discount_pct', 'revenue']
plt.figure(figsize=(8,6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
